```python
import pandas as pd
import numpy as np
```

```python
# Importar base de datos municipal y departamental
datos_municipal = pd.read_excel("guajira.xlsx", sheet_name = "1PM_arreglado", decimal = ',', thousands = '.')
datos_municipal.columns = ['departamento','municipio','poblacion','edad','total','hombre','mujer','indice_masculinidad',
                          'indice_feminidad']
datos_municipal.head()
```

```python
datos_departamental = pd.read_excel("guajira.xlsx", sheet_name = "1PD_arreglado", decimal = ',', thousands = '.')
datos_departamental.columns = ['departamento','tipo','edad','total','hombre','mujer','indice_masculinidad','indice_feminidad']
datos_departamental.head()
```

```python
# Crear variables para graficar
for i in [datos_departamental,datos_municipal]:
    i['hombre'] = -1 * i['hombre']
```

```python
# Filtrar Tipo Población en total  centro  poblado para municipio
municipal_total = datos_municipal.loc[datos_municipal['poblacion'] == 'Total'].reset_index(drop = True)
municipal_total = municipal_total.iloc[1:]
municipal_total
```

```python
# Filtrar población en Rural
municipal_rural = datos_municipal.loc[datos_municipal['poblacion'] == 'Rural'].reset_index(drop = True)
municipal_rural = municipal_rural.iloc[1:]
municipal_rural
```

```python
#Filtrar población urbano
municipal_urbano = datos_municipal.loc[datos_municipal['poblacion'] == 'Urbano'].reset_index(drop = True)
municipal_urbano = municipal_urbano.iloc[1:]
municipal_urbano
```

```python
# Filtrar departamento en Total
departamental_total = datos_departamental.loc[datos_departamental['tipo'] == 'Total'].reset_index(drop = True)
departamental_total = departamental_total.iloc[1:]
departamental_total
```

```python
# Filtrar departamento en Rural
departamental_rural = datos_departamental.loc[datos_departamental['tipo'] == 'Rural'].reset_index(drop = True)
departamental_rural = departamental_rural.iloc[1:]
departamental_rural
```

```python
# Filtrar departamento en Urbano
departamental_urbano = datos_departamental.loc[datos_departamental['tipo'] == 'Urbano'].reset_index(drop = True)
departamental_urbano = departamental_urbano.iloc[1:]
departamental_urbano
```

```python
import plotly.graph_objects as gp

particiones = [departamental_urbano, departamental_rural, departamental_total, municipal_urbano, municipal_rural,municipal_total]
titulos = ['La Guajira Población Urbana','La Guajira Población Rural','La Guajira Total',' Riohacha Población Urbana',
          'Riohacha Población Rural', 'Riohacha Población Total']
# Creating instance of the figure
for valor,titulo in zip(particiones,titulos):
    fig = gp.Figure()
  
    # Adding Male data to the figure
    fig.add_trace(gp.Bar(y= valor['edad'], x = valor['hombre'], 
                     name = 'Hombre', 
                     orientation = 'h'))
  
    # Adding Female data to the figure
    fig.add_trace(gp.Bar(y = valor['edad'], x = valor['mujer'],
                     name = 'Mujer', orientation = 'h'))
  
    # Updating the layout for our graph
    fig.update_layout(title = f'Piramide Poblacional de {titulo} 2018',
                 title_font_size = 22, barmode = 'relative',
                 bargap = 0.0, bargroupgap = 0,
                 xaxis = dict(tickvals = [-81000, -40000, -20000,-10000,-5000,-3000,
                                          0, 3000,5000,10000,20000, 40000, 81000],
                                
                              ticktext = ['81', '40', '20', '10','5','3','0', 
                                          '3','5','10''20', '40', '81'],
                                
                              title = 'Población (Miles)',
                              title_font_size = 14)
                 )
  
    fig.show()
```
